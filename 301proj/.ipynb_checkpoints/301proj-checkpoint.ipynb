```python
# Python program for solution of M Coloring 
# problem using backtracking

import random
import time
import numpy as np
import scipy as sp
from scipy import stats

def mean_confidence_interval(data, c):
    confidence=c
    a = 1.0*np.array(data)
    n = len(a)
    m, se = np.mean(a), sp.stats.sem(a)
    h = se * sp.stats.t._ppf((1+confidence)/2., n-1)
    return m-h, m+h

def random_adjacency_matrix(n):   
    matrix = [[random.randint(0, 1) for i in range(n)] for j in range(n)]

    # No vertex connects to itself
    for i in range(n):
        matrix[i][i] = 0

    # If i is connected to j, j is connected to i
    for i in range(n):
        for j in range(n):
            matrix[j][i] = matrix[i][j]

    return matrix
 
class Graph():
    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)]\
                              for row in range(vertices)]
 
    # A utility function to check if the current color assignment
    # is safe for vertex v
    def isSafe(self, v, colour, c):
        for i in range(self.V):
            if self.graph[v][i] == 1 and colour[i] == c:
                return False
        return True
     
    # A recursive utility function to solve m
    # coloring  problem
    def graphColourUtil(self, m, colour, v):
        if v == self.V:
            return True
 
        for c in range(1, m+1):
            if self.isSafe(v, colour, c) == True:
                colour[v] = c
                if self.graphColourUtil(m, colour, v+1) == True:
                    return True
                colour[v] = 0
        print("No solution.")
        return False
 
    def graphColouring(self, m):
        colour = [0] * self.V
        if self.graphColourUtil(m, colour, 0) == False: return False
        
        # Print the solution
        print ("Solution exists") # and Following are the assigned colours:")
        for c in colour:
            print (c),
        return True


# Driver Code
def proceed(run, matrixSize):
    timer = []
    for i in range(run):
        start = time.time()
        v = matrixSize
        g  = Graph(int(v))
        matrix = random_adjacency_matrix(int(v))
        g.graph = matrix
        m=3
        end = time.time()
        timer.append(end - start)
        print(g.graphColouring(m))
    #print(timer) 
    clk = str(sum(timer) / float(len(timer)))
    stddev = str(np.std(np.asarray(timer)))
    stderr = str(stats.sem(timer))
    ninty = str(mean_confidence_interval(timer, 0.9)[0]) + " - " + str(mean_confidence_interval(timer, 0.9)[1])
    nintyfive = str(mean_confidence_interval(timer, 0.95)[0]) + " - " + str(mean_confidence_interval(timer, 0.95)[1])
    
    print("average clock: " + clk) # prints average clock time
    print("standard deviation: " + stddev) # prints standard deviation
    print("standard error: " + stderr) # prints standard error
    print("90%-CL: " + ninty)
    print("95%-CL: " + nintyfive)
    

#   filename = str(run) + "final2.csv"
#    fd = open(filename,'a')
#    fd.write("\"" + str(matrixSize) + "\",\"" + clk + "\",\""+ stddev + "\",\"" + stderr + 
#             "\",\"" + ninty + "\",\"" + nintyfive + "\"\n")
#    fd.close()

  
print("How many times to run?")
run = int(input())

print("Size of adjacent matrix?")
sz = int(input())

proceed(run,sz)


#for i in range(1,500):
#    proceed(100,i) # runs 100 test case for each loop variable

 
#This code is contributed by Divyanshu Mehta

```

```python
dataset5_2 = pd.read_csv('./raw-data/5final2.csv')
x = dataset5_2.Size.apply(getLog)
y = dataset5_2.Time.apply(getLog)

denominator = x.dot(x) - x.mean() * x.sum()
m = (x.dot(y) - y.mean() * x.sum())/denominator
b = (y.mean()*x.dot(x) - x.mean()*x.dot(y))/denominator
y_pred = m*x + b

res = y - y_pred
tot = y - y.mean()

R_squared = 1 - res.dot(res)/tot.dot(tot)
R_squared = "R-squared: " + str(R_squared)

plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(20, 15))
plt.xlabel("log(size)")
plt.ylabel("log(time)")
plt.plot(x, y, 'o', color='black')
plt.plot(x,y_pred, 'r')
plt.text(2, -2, R_squared)
plt.title("5 runs per Input Size");
```
